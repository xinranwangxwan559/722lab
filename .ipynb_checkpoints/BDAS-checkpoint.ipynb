```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Obesity Data Analysis") \
    .getOrCreate()

file_path = "Estimation of obesity levels based on eating habits and physical condition - ObesityDataSet_raw_and_data_sinthetic.csv"
df = spark.read.csv(file_path, header=True, inferSchema=True)
df.show()

summary = df.describe()
summary.show()

print(df.select("NObeyesdad").show())
```

```python
from pyspark.sql.functions import mean, stddev, abs,col
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Numerical attributes to be checked
numeric_columns = ["Age", "Height", "Weight", "FCVC", "NCP", "CH2O", "FAF", "TUE"]

outliers_dict_zscore = {}

for column in numeric_columns:
    # Calculate mean and standard deviation
    avg = df.agg(mean(col(column))).first()[0]
    std_dev = df.agg(stddev(col(column))).first()[0]
    
    # calculate Z-score
    zscore_expr = (col(column) - avg) / std_dev
    zscore_col = abs(zscore_expr).alias("zscore")
    
    # Filtering outliers
    outliers = df.withColumn("zscore", zscore_col).filter("zscore > 3 or zscore < -3")
    
    # Putting Exceptions into a Dictionary
    outliers_dict_zscore[column] = outliers

# Output the number of outliers for each attribute
for column, outlier_df in outliers_dict_zscore.items():
    print(f"{column} has {outlier_df.count()} outliers.")
    


# transfer Pandas DataFrame
pandas_df = df.toPandas()

# set Seaborn 
sns.set(style="whitegrid")

#  Boxplot
plt.figure(figsize=(15, 10))
sns.boxplot(data=pandas_df[numeric_columns])

# print name 
plt.title("Boxplot of Numeric Attributes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(df.select("NObeyesdad").show())
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import monotonically_increasing_id, when, col

# Initialize a Spark session
spark = SparkSession.builder.appName("MergeDatasets").getOrCreate()

# Read the original dataset
df = spark.read.csv("Estimation of obesity levels based on eating habits and physical condition - ObesityDataSet_raw_and_data_sinthetic.csv", header=True, inferSchema=True)

# Read the new dataset
new_df = spark.read.csv("obesity classification - Obesity Classification.csv", header=True, inferSchema=True)

# Add ID columns to the original and new datasets
df = df.withColumn("ID", monotonically_increasing_id())
new_df = new_df.withColumn("ID", monotonically_increasing_id())

# Calculate BMR_Estimate for df
bmr_male = 88.362 + (13.397 * col("Weight")) + (4.799 * col("Height") * 100) - (5.677 * col("Age"))
bmr_female = 447.593 + (9.247 * col("Weight")) + (3.098 * col("Height") * 100) - (4.330 * col("Age"))
df = df.withColumn("BMR_Estimate", when(col("Gender") == "Male", bmr_male).otherwise(bmr_female))

# Add Age_Group to df
df = df.withColumn(
    "Age_Group",
    when(col("Age").between(14, 18), "14-18")
    .when(col("Age").between(19, 24), "19-24")
    .when(col("Age").between(25, 34), "25-34")
    .when(col("Age").between(35, 44), "35-44")
    .when(col("Age").between(45, 54), "45-54")
    .when(col("Age").between(55, 64), "55-64")
    .otherwise("65+")
)

# Ensure new_df does not have Age_Group and BMR_Estimate columns
if "Age_Group" in new_df.columns:
    new_df = new_df.drop("Age_Group")

if "BMR_Estimate" in new_df.columns:
    new_df = new_df.drop("BMR_Estimate")

# Merge the dataframes
merged_df = df.join(new_df, "ID", "outer")

# Display the merged dataframe columns
print(merged_df.columns)

# Display the merged data set
merged_df.show()
merged_df.columns
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, monotonically_increasing_id

spark = SparkSession.builder.appName("CalculateAvgValues").getOrCreate()

# rename df columns
df = df.withColumnRenamed("Age", "df_Age")\
       .withColumnRenamed("Height", "df_Height")\
       .withColumnRenamed("Weight", "df_Weight")

# rename new df columns
new_df = new_df.withColumnRenamed("Age", "new_df_Age")\
               .withColumnRenamed("Height", "new_df_Height")\
               .withColumnRenamed("Weight", "new_df_Weight")

# add ID
df = df.withColumn("ID", monotonically_increasing_id())
new_df = new_df.withColumn("ID", monotonically_increasing_id())

# merge data
merged_df = df.join(new_df, "ID", "outer")

# calculate avg
merged_df = merged_df.withColumn("Avg_Age", (col("df_Age") + col("new_df_Age")) / 2)
merged_df = merged_df.withColumn("Avg_Height", (col("df_Height") + col("new_df_Height")) / 2)
merged_df = merged_df.withColumn("Avg_Weight", (col("df_Weight") + col("new_df_Weight")) / 2)

# show result
merged_df.select("Avg_Age", "Avg_Height", "Avg_Weight").show()

# show description
summary = merged_df.describe()
summary.show()
```

```python
from pyspark.ml.feature import ChiSqSelector, VectorAssembler, StringIndexer

# Define all non-numeric columns to be converted
non_numeric_cols = ["Gender", "family_history_with_overweight", "FAVC", "CAEC", "SMOKE", "SCC", "CALC", "MTRANS", "Age_Group"]

# Converting non-numeric columns with StringIndexer
indexers = []
for col in non_numeric_cols:
    indexer = StringIndexer(inputCol=col, outputCol=col+"_encoded")
    indexers.append(indexer)

# Coding of the target column NObeyesdad
label_indexer = StringIndexer(inputCol="NObeyesdad", outputCol="NObeyesdad_encoded")
indexers.append(label_indexer)

# Use Pipeline to perform all conversions at once
from pyspark.ml import Pipeline
pipeline = Pipeline(stages=indexers)
df_encoded = pipeline.fit(df).transform(df)

# Get all columns for feature vectorization
feature_cols = df_encoded.columns
feature_cols.remove("NObeyesdad")

# Update feature columns to include only coded columns and other numeric columns
encoded_feature_cols = [col+"_encoded" for col in non_numeric_cols]
numeric_cols = list(set(feature_cols) - set(non_numeric_cols))
all_feature_cols = numeric_cols + encoded_feature_cols

# Combining Features with VectorAssembler
assembler = VectorAssembler(inputCols=all_feature_cols, outputCol="features")
df_vectorized = assembler.transform(df_encoded)

# Feature Selection with ChiSqSelector
selector = ChiSqSelector(numTopFeatures=10, featuresCol="features", outputCol="selectedFeatures", labelCol="NObeyesdad_encoded")
df_selected = selector.fit(df_vectorized).transform(df_vectorized)

# Show selected features
df_selected.select("selectedFeatures").show()

# Get the index of the selected feature
selected_indices = selector.fit(df_vectorized).selectedFeatures

# Use these indexes to extract feature names from all_feature_cols
selected_feature_names = [all_feature_cols[index] for index in selected_indices]

print("Selected features:")
for feature in selected_feature_names:
    print(feature)

```

```python
from pyspark.sql.functions import sqrt
from pyspark.ml.feature import OneHotEncoder, StringIndexer
from pyspark.ml import Pipeline

# Properties to check
selected_features = ["NObeyesdad_encoded", "FCVC", "Age_Group_encoded", "CAEC_encoded", "MTRANS_encoded", "df_Weight", "FAVC_encoded", "Gender_encoded", "SCC_encoded", "family_history_with_overweight_encoded"]

# Apply sqrt transformation
for feature in selected_features:
    df_selected = df_selected.withColumn(feature, sqrt(df_selected[feature]))

# Using Pipeline for OneHotEncoder
stages = []
for feature in selected_features:
    # StringIndexer converts string categorization tags to label indicators
    stringIndexer = StringIndexer(inputCol=feature, outputCol=feature + "_indexed")
    
    # OneHotEncoder converts labeled metrics into one-hot encoded vectors
    encoder = OneHotEncoder(inputCol=feature + "_indexed", outputCol=feature + "_vec")
    
    stages += [stringIndexer, encoder]

pipeline = Pipeline(stages=stages)
df_transformed = pipeline.fit(df_selected).transform(df_selected)

# Display the transformed data
df_transformed.select([f + "_vec" for f in selected_features]).show()

# Plotting histograms 
import matplotlib.pyplot as plt
for feature in selected_features:
    # Extraction of feature data
    data = df_selected.select(feature).rdd.flatMap(lambda x: x).collect()
    
    # Plotting histograms
    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=50, color='blue', alpha=0.7)
    plt.title(f'Histogram for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
from pyspark.sql.functions import sqrt, col
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# feature selection
selected_features = ["NObeyesdad_encoded", "FCVC", "Age_Group_encoded", "CAEC_encoded", "MTRANS_encoded", "df_Weight", "FAVC_encoded", "Gender_encoded", "SCC_encoded", "family_history_with_overweight_encoded"]

# Apply square root conversion to non-target columns only
for feature in selected_features:
    if feature != "NObeyesdad_encoded":
        df_selected = df_selected.withColumn(feature, sqrt(df_selected[feature]))

# Convert NObeyesdad_encoded to integer
df_selected = df_selected.withColumn("NObeyesdad_encoded", col("NObeyesdad_encoded").cast("int"))

# Using Pipeline for OneHotEncoder
stages = []
for feature in selected_features:
    stringIndexer = StringIndexer(inputCol=feature, outputCol=feature + "_indexed")
    encoder = OneHotEncoder(inputCol=feature + "_indexed", outputCol=feature + "_vec")
    stages += [stringIndexer, encoder]

pipeline = Pipeline(stages=stages)
df_transformed = pipeline.fit(df_selected).transform(df_selected)

# Delete the "features" column if it already exists.
if "features" in df_transformed.columns:
    df_transformed = df_transformed.drop("features")

# Combining Features with VectorAssembler
feature_columns = [f + "_vec" for f in selected_features]
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df_transformed = assembler.transform(df_transformed)

# Training a Decision Tree Classifier
dt = DecisionTreeClassifier(labelCol="NObeyesdad_encoded", featuresCol="features")
dt_model = dt.fit(df_transformed)
predictions = dt_model.transform(df_transformed)

# View the structure of the decision tree
print(dt_model.toDebugString)

# Evaluating Model Accuracy with MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(
    labelCol="NObeyesdad_encoded", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Show forecast results
predictions.select("NObeyesdad_encoded", "prediction", "probability").show(truncate=False)
```

```python
from pyspark.sql.functions import corr, countDistinct, mean, stddev

# 10
selected_features = ['FAVC_encoded', 'SCC_encoded', 'CALC_encoded', 'df_Age', 'MTRANS_encoded', 'FCVC', 'df_Weight', 'Age_Group_encoded', 'Gender_encoded', 'NObeyesdad_encoded']
selected_data = df_transformed.select(*selected_features)

# 
for feature in selected_features[:-1]:  # NObeyesdad_encoded
    # NObeyesdad_encoded
    correlation = selected_data.stat.corr(feature, 'NObeyesdad_encoded')
    print(f"Correlation between {feature} and NObeyesdad_encoded: {correlation:.2f}")
    
    # 
    unique_values = selected_data.agg(countDistinct(feature)).collect()[0][0]
    if unique_values < 10:
        print(f"{feature} has less than 10 unique values, suggesting it might be categorical or ordinal in nature.")
    else:
        print(f"{feature} seems continuous with {unique_values} unique values.")
    
    # 
    feature_mean = selected_data.agg(mean(feature)).collect()[0][0]
    feature_stddev = selected_data.agg(stddev(feature)).collect()[0][0]
    if feature_stddev:
        skewness = 3 * (feature_mean - feature_stddev) / feature_stddev
        if skewness > 1:
            print(f"{feature} seems to be right-skewed.")
        elif skewness < -1:
            print(f"{feature} seems to be left-skewed.")
        else:
            print(f"{feature} seems to be symmetrically distributed.")
    else:
        print(f"{feature} has no variability.")
    
    print("\n")
```

```python
from pyspark.ml.classification import LinearSVC, OneVsRest
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Randomly split the data into a training set and a test set
(train_data, test_data) = df_transformed.randomSplit([0.8, 0.2])

# Initialize the LinearSVC classifier with specified parameters
lsvc = LinearSVC(labelCol="NObeyesdad_encoded", featuresCol="features", maxIter=10, standardization=True)

# Using the OneVsRest classifier
ovr = OneVsRest(classifier=lsvc, labelCol="NObeyesdad_encoded", featuresCol="features")

# Set up the parameter grid
paramGrid = ParamGridBuilder() \
    .addGrid(lsvc.regParam, [0.01, 0.1, 0.5, 1.0, 2.0]) \
    .build()

# Create train-validation split to determine best model
tvs = TrainValidationSplit(estimator=ovr,
                           estimatorParamMaps=paramGrid,
                           evaluator=MulticlassClassificationEvaluator(labelCol="NObeyesdad_encoded", predictionCol="prediction", metricName="accuracy"),
                           trainRatio=0.8)

# Fit model to train_data and choose the best set of parameters
best_model = tvs.fit(train_data)

# Using the best model to predict test set data
predictions = best_model.transform(test_data)

# Show forecast results
predictions.select("NObeyesdad_encoded", "prediction").show(truncate=False)

# Using MulticlassClassificationEvaluator to Evaluate Model Performance on the Test Set
evaluator = MulticlassClassificationEvaluator(labelCol="NObeyesdad_encoded", predictionCol="prediction")

# Calculation metrics
accuracy = evaluator.setMetricName("accuracy").evaluate(predictions)
print(f"Test Accuracy: {accuracy:.2f}")

weighted_precision = evaluator.setMetricName("weightedPrecision").evaluate(predictions)
print(f"Test Weighted Precision: {weighted_precision:.2f}")

weighted_recall = evaluator.setMetricName("weightedRecall").evaluate(predictions)
print(f"Test Weighted Recall: {weighted_recall:.2f}")

f1_score = evaluator.setMetricName("f1").evaluate(predictions)
print(f"Test F1 Score: {f1_score:.2f}")

# Print out the best C-Parameter for LinearSVC
print(f"Best C-Parameter: {lsvc.getRegParam()}")
```
